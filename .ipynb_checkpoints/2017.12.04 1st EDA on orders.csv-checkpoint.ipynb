## orders['days_since_prior_order'] 에 존재하는 nan값 처리
* orders['days_since_prior_order'] 에는 nan 값이 존재한다. nan값은 처음 구매한 시점으로 볼수 있는데 0으로 변환 할수 있을까?
* orders['days_since_prior_order'] 값이 이미 0인 케이스가 존재하다. (무수히 많다)
* orders['days_since_prior_order'] 값이 0인 케이스는 user가 구매시 어떤 물품구매를 하지못해 바로 같은 날다시 구매를 했을때임 을 알 수 있다
* 데이터 orders['order_dow'] 가 같을 때 후자의 데이터 orders['days_since_prior_order'] 값이 0임을 확인 할 수 있다
* 위의 statement는 거꾸로 생각 해 보았을 땐, 성립되지 않는다 (요일의 값이므로 다른 주 같은 요일일 땐, orders['days_since_prior_order']는 0 이 될 수 없다 
* 그러므로 orders['days_since_prior_order'] 의 nan 값을 0으로 변환 시킬 수 "없다"

#### 한가지 방법으로 nan 값을 처리 할수 있다
우선 orders['days_since_prior_order'] 의 nan 값을 -1로 표현한 후, 모든 값들을 다시 categorize 해 줄수 있다
* orders['days_since_prior_order'] 값이 = 0 은 사실상 같은 날 여러번의 구매이지만, 그 날 가장 일찍 일어난 값으로 바꿔 줄 수 없다.
* orders['days_since_prior_order'] 데이터 자체가 전 구매로부터 day의 차이로 결정되는 값이으로 그 후 일어난 구매들의 값이 영향을 받는다
* 소비자의 입장에선 하나의 장바구니에서 처리되었어야 할 구매에 실수가 일어났다기 때문에, 하나의 구매로 볼 수 있지만, 위와 같은 이유로 각각 다른 구매로 보는 것이 마땅하다


#### 전 구매 날과 다음 구매 날까지의 간격
* 7일 후 와 30일 후가 가장 높은 것으로 발견됨
* 30일이 높은 이유는 말그대로 30일 간격으로 구매가 많을수도 있고, 30일 후의 모든 구매간격을 30일로 통합 시켰을 수도 있음

```python
# 위에 대한 증명의 간단한 예
for i in range(len(orders['days_since_prior_order'])):
    if orders['days_since_prior_order'][i] == 0:
        if i <= 10:
            print (orders['user_id'][i])
```

```python
orders['days_since_prior_order'] = orders['days_since_prior_order'].replace(np.nan, -1)
```

```python
orders['days_since_prior_order'].head(10)
```

```python
def encode_(x):
    for i in range(30):
        if x == -1: return 'first order'
        elif x <= 1: return 'within {} day(s)'.format(1)
        elif x <= i: return 'within {} day(s)'.format(i)
        elif x == 30: return 'within {} day(s)'.format(30)

orders['days_since_prior_order'] = orders['days_since_prior_order'].apply(encode_)
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation='vertical')
sns.countplot(orders['days_since_prior_order'])
```
